### Experiments

The experiments ran are fairly straightfoward. The dataset is sampled and split into a training (70%) and testing (30%) dataset where both models, ID3 and KM, are trained and tested on the same groups. Training wise, we generate two mdoels. One model tis the decision tree by the ID3 algorithm. The second model is a set of clusters built by running the KM algorithm. The second stage of the experiment consisits of testing. The genearted models from the `train` step are read and used to classify several datasets.

```python
%matplotlib inline
import seaborn as sns
import pandas as pd

compare = pd.read_csv("data/compare.csv")

sns.violinplot(
    x="algorithm",
    y="predicted",
    hue="correct",
    data=compare,
    split=True,
    inner="stick",
    palette="Set3"
)

sns.plt.show()
```

No, not enough graphs yet. Let's add another one and talk more about stuff!

```python
sns.factorplot(
    x="correct",
    data=compare,
    palette="pastel",
    col="algorithm",
    hue="predicted",
    kind="count"
)

sns.plt.show()
```

And, one last plot for the road.

```python
sns.factorplot(
    x="predicted",
    y="actual",
    data=compare,
    palette="deep",
    kind="swarm",
    col="algorithm"
)

sns.plt.show()
```
